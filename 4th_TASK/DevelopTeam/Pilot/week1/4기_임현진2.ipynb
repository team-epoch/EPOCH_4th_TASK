```python
import pandas as pd
csv_path = '/content/diabetes2.csv'
data = pd.read_csv(csv_path, encoding='latin-1')
```

```python
data
```

```python
X = data[['Pregnancies', 'BMI', 'Age']]
y = data['BloodPressure' ]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
pred= model.predict(X_test)
comparison= pd.DataFrame({'Actual': y_test, 'Predicted': pred})
comparison
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.scatterplot(x='Actual', y='Predicted', data= comparison)
```

```python
model.score(X_test, y_test)
```

해당 모델은 결정계수가 약 0.15 정도로 좋은 모델이라고 볼 수 없다. 독립변수(Pregnancies, BMI, Age)가 종속변수(BloodPressure)를 15% 정도 설명하고 있다고 볼 수 있다. 연습 삼아 좋은 모델을 더 만들어 보고 싶다.
