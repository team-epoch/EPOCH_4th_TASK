```python
import pandas as pd

shots_df=pd.read_csv('processing_shots_df.csv')

shots_df.head()
```

```python
# ëª¨ë¸ í•™ìŠµì— ì‚¬ìš©í•  ë³€ìˆ˜
features = [
    "X", "Y",
    "shot_distance", "shot_angle",
    "assist_flag", "home_away",
    "situation_FromCorner", "situation_OpenPlay",
    "situation_Penalty", "situation_SetPiece",
    "shotType_LeftFoot", "shotType_RightFoot", "shotType_OtherBodyPart",
    "lastAction_encoded"
]
original_xG=shots_df['xG'] # ì‹¤ì œ xGê°’
target = "is_goal"
```

```python
shots_df[features].describe()
```

```python
from sklearn.model_selection import train_test_split

X = shots_df[features]
y = shots_df[target]

X_train, X_test, y_train, y_test, xG_train, xG_test = train_test_split(
    X, y, original_xG, test_size=0.2, random_state=42
) # xGê°’ì„ ë‚˜ì¤‘ì— ë¹„êµí•˜ê¸° ìœ„í•´ ê°™ì€ ì¸ë±ìŠ¤ë¡œ ë¶„ë¦¬í•´ë‘ 
```

```python
print("í•™ìŠµ ë°ì´í„° í˜•íƒœ:", X_train.shape)
print("í…ŒìŠ¤íŠ¸ ë°ì´í„° í˜•íƒœ:", X_test.shape)
```

```python
import pandas as pd
import numpy as np

# ëª¨ë¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import lightgbm as lgb
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score


X_train, X_test, y_train, y_test, xG_train, xG_test = train_test_split(
    X, y, original_xG, test_size=0.2, random_state=42
)


# ì—¬ëŸ¬ ëª¨ë¸ ì •ì˜ 
models = {
    "LightGBM": lgb.LGBMClassifier(random_state=42),
    "XGBoost": xgb.XGBClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
}


# ëª¨ë¸ í•™ìŠµ ë° ì„±ëŠ¥ ë¹„êµ
# ê° ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ì €ì¥í•  ë”•ì…”ë„ˆë¦¬
results = {}

for name, model in models.items():
    print(f"--- {name} ëª¨ë¸ í•™ìŠµ ì¤‘... ---")
    
    # ëª¨ë¸ í•™ìŠµ
    model.fit(X_train, y_train)
    
    # xG ì˜ˆì¸¡ (ê³¨ì´ ë  í™•ë¥ )
    predicted_xG = model.predict_proba(X_test)[:, 1]
    
    # ì„±ëŠ¥ í‰ê°€ (ì‹¤ì œ xG ê°’ê³¼ ë¹„êµ)
    mae = mean_absolute_error(xG_test, predicted_xG)
    r2 = r2_score(xG_test, predicted_xG)
    
    # ê²°ê³¼ ì €ì¥
    results[name] = {'MAE': mae, 'R2': r2}
    print(f"{name} ëª¨ë¸ í‰ê°€ ì™„ë£Œ: MAE={mae:.4f}, R2={r2:.4f}\n")


# ìµœì¢… ê²°ê³¼ ë¹„êµ
print("--- ìµœì¢… ëª¨ë¸ ì„±ëŠ¥ ë¹„êµ ---")
results_df = pd.DataFrame(results).T # .TëŠ” í–‰ê³¼ ì—´ì„ ë°”ê¿ˆ (Transpose)
print(results_df.sort_values(by='R2', ascending=False)) # R2ê°€ ë†’ì€ ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ì¶œë ¥
```

- ê° ë³€ìˆ˜ì™€ ê²°ê³¼ ì‚¬ì´ê°€ ì„ í˜•ì´ë¼ ê°€ì •í•˜ëŠ” í•œê³„ê°€ ìˆìŒ
- ë¹„ì„ í˜• ê´€ê³„ë¥¼ í•™ìŠµí•˜ê¸° ìœ„í•´ì„œëŠ” ì§ì ‘ í”¼ì³ ì—”ì§€ë‹ˆì–´ë§ì„ í•´ì¤˜ì•¼í•¨

```python
model = LogisticRegression()
model.fit(X_train, y_train)


# íšŒê·€ ê³„ìˆ˜ í™•ì¸ (ê°€ì¥ ê¸°ë³¸ì ì¸ ë°©ë²•)
print("--- ê¸°ë³¸ ì¶œë ¥ (Numpy ë°°ì—´) ---")
print(model.coef_)
print("\n" + "="*30 + "\n")


# í”¼ì²˜ ì´ë¦„ê³¼ í•¨ê»˜ ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥ (ì¶”ì²œ)
# pandas Seriesë¥¼ í™œìš©í•˜ì—¬ í”¼ì²˜ ì´ë¦„ê³¼ ê³„ìˆ˜ë¥¼ ì§ì§€ì–´ì¤ë‹ˆë‹¤.
coefficients = pd.Series(model.coef_[0], index=X_train.columns)

print("--- í”¼ì²˜ë³„ íšŒê·€ ê³„ìˆ˜ ---")
print(coefficients.sort_values(ascending=False)) # ì˜í–¥ë ¥ì´ í° ìˆœìœ¼ë¡œ ì •ë ¬

# ì ˆí¸(intercept) í™•ì¸
print("\n--- ì ˆí¸ (Intercept) ---")
print(model.intercept_[0])
```

## ë¶€ì •ì  ì˜í–¥ (ê³¨ í™•ë¥  ê°ì†Œ ìš”ì¸) ğŸ“‰

ê³„ìˆ˜ê°€ ìŒìˆ˜(-)ì¸ í”¼ì²˜ë“¤ì€ ê³¨ í™•ë¥ ì„ ë‚®ì¶”ëŠ” ìš”ì†Œ. ìˆ«ìì˜ ì ˆëŒ“ê°’ì´ í´ìˆ˜ë¡ ì˜í–¥ë ¥ì´ ë” í¼.

- shot_distance (-2.2181): ê°€ì¥ ê°•ë ¥í•œ ë¶€ì • ìš”ì¸. ê³¨ë¬¸ê³¼ì˜ ê±°ë¦¬ê°€ ë©€ì–´ì§ˆìˆ˜ë¡ ê³¨ í™•ë¥ ì´ ê¸‰ê²©í•˜ê²Œ ë‚®ì•„ì§„ë‹¤ëŠ”, ì¶•êµ¬ì˜ ê°€ì¥ ê¸°ë³¸ì ì¸ ìƒì‹ì„ ëª¨ë¸ì´ ì •í™•í•˜ê²Œ ë°˜ì˜.

- situation_FromCorner (-0.4767): ì½”ë„ˆí‚¥ìœ¼ë¡œë¶€í„° ì´ì–´ì§„ ìŠ›ì€ ë‹¤ë¥¸ ìƒí™©(ì˜¤í”ˆ í”Œë ˆì´ ë“±)ì— ë¹„í•´ ê³¨ì´ ë  í™•ë¥ ì´ ë‚®ìŒ. ì½”ë„ˆí‚¥ ìƒí™©ì´ í˜¼ì „ ì†ì—ì„œ ë‚˜ì˜¤ëŠ” ì§ˆ ë‚®ì€ ìŠˆíŒ…ì´ ë§ë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸.

```python
from sklearn.model_selection import KFold

logit_model = LogisticRegression(random_state=42)

# êµì°¨ê²€ì¦ ì„¤ì •
cv = KFold(n_splits=5, shuffle=True, random_state=42)

mae_scores, r2_scores = [], []

# Foldë³„ í•™ìŠµ & í‰ê°€
for train_idx, test_idx in cv.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    xG_test = original_xG.iloc[test_idx]   # <- ë¹„êµ ëŒ€ìƒì€ original_xG
    
    logit_model.fit(X_train, y_train)
    
    predicted_xG = logit_model.predict_proba(X_test)[:, 1]
    
    # original_xGì™€ ë¹„êµ 
    mae = mean_absolute_error(xG_test, predicted_xG)
    r2 = r2_score(xG_test, predicted_xG)
    
    mae_scores.append(mae)
    r2_scores.append(r2)

logit_cv_summary = {
    "MAE_mean": np.mean(mae_scores),
    "R2_mean": np.mean(r2_scores),
}

print("--- Logistic Regression êµì°¨ê²€ì¦ ê²°ê³¼ ---")
print(pd.DataFrame([logit_cv_summary]))
```
