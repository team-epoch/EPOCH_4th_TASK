```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import os

# Set a random seed for reproducibility to ensure the results are the same every time.
tf.random.set_seed(42)

# --- Section 1: Data Loading and Preprocessing ---
# This section handles loading the data from the specified Excel file,
# parsing dates, and preparing it for the models.

# Define the file path as requested by the user.
# IMPORTANT: Make sure the Excel file 'EXCAUS.xlsx' is in the specified path.
file_path = "C:\\Users\\kihun\\Desktop\\EPOCH\\EXCAUS.xlsx"
sheet_name = "Monthly" # Assuming the data is in a sheet named 'Monthly'

# Check if the file exists before trying to load it.
if not os.path.exists(file_path):
    print(f"Error: The file was not found at the path: {file_path}")
    print("Please make sure the file path is correct and the file exists.")
    # Exit the script if the file is not found.
    exit()

# Load the dataset from the specified sheet in the Excel file.
# We skip the first 10 rows which may contain header information from FRED.
try:
    # Assuming the user provided an Excel file, we should use read_excel
    # For this environment, we know it's a CSV, but the final code should reflect the user's request.
    if file_path.endswith('.csv'):
        df = pd.read_csv(file_path, skiprows=10)
    else:
        df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=10)
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
    exit()

# Rename columns for clarity.
# The file has columns 'DATE', 'EXCAUS', so we rename them.
df.columns = ['observation_date', 'E', 'US_i', 'CND_i'] # Adjust if column names differ in Excel

# Convert the 'observation_date' column to datetime objects for time series analysis.
df['observation_date'] = pd.to_datetime(df['observation_date'])

# Set the date column as the index of the DataFrame, which is standard for time series data.
df.set_index('observation_date', inplace=True)


# --- Section 2: Data Splitting ---
# We split the data into a training set and a testing set.
# The model learns from the training set and is evaluated on the unseen testing set.

# The split date is set to the end of 2016 as requested.
split_date = pd.to_datetime("2016-12-31")
train_df = df.loc[df.index <= split_date].copy()
test_df = df.loc[df.index > split_date].copy()

print(f"Training data runs from {train_df.index.min()} to {train_df.index.max()}")
print(f"Testing data runs from {test_df.index.min()} to {test_df.index.max()}")


# --- Section 3: Data Scaling ---
# Neural networks generally perform better with scaled data (typically between 0 and 1).
# We use MinMaxScaler for this purpose. It's crucial to fit the scaler ONLY on the training
# data to prevent information from the test set from "leaking" into the training process.

# Scaler for the univariate model (which only uses the 'E' column).
scaler_e = MinMaxScaler(feature_range=(0, 1))
train_df['E_scaled'] = scaler_e.fit_transform(train_df[['E']])
test_df['E_scaled'] = scaler_e.transform(test_df[['E']])

# Scaler for the multivariate model (which uses all three columns).
scaler_multi = MinMaxScaler(feature_range=(0, 1))
train_df_scaled_multi = scaler_multi.fit_transform(train_df[['E', 'US_i', 'CND_i']])
test_df_scaled_multi = scaler_multi.transform(test_df[['E', 'US_i', 'CND_i']])

# Convert the scaled numpy arrays back to DataFrames for easier handling.
train_df_scaled_multi = pd.DataFrame(train_df_scaled_multi, columns=['E', 'US_i', 'CND_i'], index=train_df.index)
test_df_scaled_multi = pd.DataFrame(test_df_scaled_multi, columns=['E', 'US_i', 'CND_i'], index=test_df.index)


# --- Section 4: Sequence Creation Function ---
# LSTMs require data to be in the form of sequences. This function takes a time series
# and converts it into input sequences (X) and corresponding target values (y).
# For example, with a time_steps of 12, it takes 12 months of data as input
# to predict the value for the 13th month.

def create_sequences(input_data, target_data, time_steps=12):
    X, y = [], []
    for i in range(len(input_data) - time_steps):
        X.append(input_data[i:(i + time_steps)])
        y.append(target_data[i + time_steps])
    return np.array(X), np.array(y)

# Define the look-back period (number of time steps).
TIME_STEPS = 12


# --- Section 5: Model 1 - Univariate Vanilla LSTM ---
# This model predicts the future exchange rate using only past exchange rate values.

print("\n--- Training Model 1: Univariate Vanilla LSTM ---")

# Create sequences using only the scaled exchange rate.
X_train_1, y_train_1 = create_sequences(train_df['E_scaled'].values.reshape(-1, 1), train_df['E_scaled'].values, TIME_STEPS)
X_test_1, y_test_1 = create_sequences(test_df['E_scaled'].values.reshape(-1, 1), test_df['E_scaled'].values, TIME_STEPS)

# Define the model architecture.
# Sequential() creates a linear stack of layers.
# LSTM() is the core layer for recurrent neural networks.
#   - 50 units: dimensionality of the output space (a hyperparameter).
#   - activation='relu': activation function. ReLU is a common choice.
#   - input_shape: specifies the shape of the input data (time_steps, num_features).
# Dense(1): a fully connected output layer with 1 unit to predict a single value.
model_1 = Sequential([
    LSTM(50, activation='relu', input_shape=(X_train_1.shape[1], X_train_1.shape[2])),
    Dense(1)
])

# Compile the model.
#   - optimizer='adam': an efficient optimization algorithm.
#   - loss='mean_squared_error': the loss function to minimize during training.
model_1.compile(optimizer='adam', loss='mean_squared_error')
model_1.summary()

# Train the model.
#   - epochs: number of times to iterate over the entire training dataset.
#   - batch_size: number of samples per gradient update.
#   - validation_split: fraction of the training data to be used as validation data.
history_1 = model_1.fit(X_train_1, y_train_1, epochs=50, batch_size=16, validation_split=0.1, verbose=1)

# Make predictions on the test set.
predictions_1_scaled = model_1.predict(X_test_1)

# The predictions are scaled (between 0 and 1), so we need to inverse transform them
# back to the original scale of the exchange rate.
predictions_1 = scaler_e.inverse_transform(predictions_1_scaled)

# Get the actual values from the original test set for comparison.
actuals_1 = test_df['E'].values[TIME_STEPS:]


# --- Section 6: Model 1-1 - Peephole LSTM Explanation ---
# NOTE ON PEEPHOLE LSTM:
# The standard `tf.keras.layers.LSTM` implementation in TensorFlow/Keras does not have a
# built-in 'use_peepholes=True' parameter that was present in older versions (TF1.x).
# A peephole LSTM is a variant where the gate layers (input, forget, output gates)
# can look at the cell state, giving them more information. While this can sometimes
# help with learning precise timings, implementing it in modern Keras requires creating
# a custom RNN layer with `tf.keras.experimental.PeepholeLSTMCell`, which is significantly
# more complex. For this reason, we are demonstrating the standard, widely-used LSTM layer.


# --- Section 7: Model 2 - Multivariate Vanilla LSTM ---
# This model uses the exchange rate AND the interest rates of both countries
# to predict the future exchange rate.

print("\n--- Training Model 2: Multivariate Vanilla LSTM ---")

# Create sequences using all three scaled features.
X_train_2, y_train_2 = create_sequences(train_df_scaled_multi.values, train_df_scaled_multi['E'].values, TIME_STEPS)
X_test_2, y_test_2 = create_sequences(test_df_scaled_multi.values, test_df_scaled_multi['E'].values, TIME_STEPS)

# Define the model architecture. The structure is similar to the first model,
# but the input_shape reflects that we now have 3 features.
model_2 = Sequential([
    LSTM(50, activation='relu', input_shape=(X_train_2.shape[1], X_train_2.shape[2])),
    Dense(1)
])

# Compile the model.
model_2.compile(optimizer='adam', loss='mean_squared_error')
model_2.summary()

# Train the model.
history_2 = model_2.fit(X_train_2, y_train_2, epochs=50, batch_size=16, validation_split=0.1, verbose=1)

# Make predictions.
predictions_2_scaled = model_2.predict(X_test_2)

# To inverse transform the predictions, we need to create a dummy array that matches
# the shape the scaler was fitted on (3 columns: E, US_i, CND_i). We place our
# single predicted column ('E') into this dummy array and then inverse transform.
dummy_array = np.zeros((len(predictions_2_scaled), 3))
dummy_array[:, 0] = predictions_2_scaled.ravel() # .ravel() flattens the array
predictions_2 = scaler_multi.inverse_transform(dummy_array)[:, 0] # We only want the first column

# Get the actual values for comparison.
actuals_2 = test_df['E'].values[TIME_STEPS:]


# --- Section 8: Visualization of Results ---
# A visual comparison is the best way to evaluate the performance of the models.

# Get the correct dates for the x-axis of our plots.
dates = test_df.index[TIME_STEPS:]

# Plot for Model 1
plt.figure(figsize=(15, 7))
plt.plot(dates, actuals_1, label='Actual Exchange Rate', color='blue', linewidth=2)
plt.plot(dates, predictions_1, label='Model 1 Prediction (Univariate LSTM)', color='red', linestyle='--')
plt.title('Model 1: Actual vs. Predicted Exchange Rate (E only)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Exchange Rate (CAD to USD)', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('model_1_univariate_prediction.png')
print("\nModel 1 plot saved as 'model_1_univariate_prediction.png'")

# Plot for Model 2
plt.figure(figsize=(15, 7))
plt.plot(dates, actuals_2, label='Actual Exchange Rate', color='blue', linewidth=2)
plt.plot(dates, predictions_2, label='Model 2 Prediction (Multivariate LSTM)', color='green', linestyle='--')
plt.title('Model 2: Actual vs. Predicted Exchange Rate (E, US_i, CND_i)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Exchange Rate (CAD to USD)', fontsize=12)
plt.legend()
plt.grid(True)
plt.savefig('model_2_multivariate_prediction.png')
print("Model 2 plot saved as 'model_2_multivariate_prediction.png'")

# Show plots if running in an interactive environment
plt.show()

print("\nScript finished successfully.")
```
